### Автомобілі на якому пальному використовувалися протягом даного нам проміжку часу?

Використання ліній дасть нам зрозуміти загальні тенденції по створенню/використанню моделей автомобілів на певному типі пального. Оскільки pure gasoline моделі автомобілів значно переважають будь-яку іншу категорію, я добавив опцію interactive(), щоб можна було заскейлитись в нижню частину графіка для порівняння категорій з меншості.

З графіку спостерігається тренд росту кількості електроавтомобілів з ~2011 року.

```python
alt.Chart(df2).mark_line(point=alt.OverlayMarkDef()).encode(
    x=alt.X("year:O"),
    y=alt.Y("make", title='Model count', scale = alt.Scale(domain = [0, df2.make.max()])),
    color=alt.Color("fuel_type_1:N", title="Fuel type"),
).properties(
   width=800, height=500
).interactive(bind_x=False)
```

Також одна з цікавих тенденцій в наш час - автоматична коробка передач практично повністю витісняє механічну на ринку. Я хотів візуалізувати та й для себе дізнатися в яких саме числах/пропорціях випускалися різні типи.|

```python
df_transmission = df.drop(df.columns.difference(['model', 'year', 'trany']), 1)
df_transmission['trany'] = df_transmission['trany'].str.split().str[0]
df_transmission
```

```python
df_transmission = df_transmission.groupby(['trany', 'year']).count().reset_index()
df_transmission
```

Також одна з цікавих тенденцій в наш час - автоматична коробка передач практично повністю витісняє механічну на ринку. Я хотів візуалізувати та й для себе дізнатися в яких саме числах/пропорціях випускалися різні типи.

Даний графік дозволяє зразу побачити загальну тенденцію та порівняти обидва значення кожного року, навівши на лінію курсор.

```python
nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['year'], empty='none')

line = alt.Chart(df_transmission).mark_line(interpolate='basis').encode(
    x=alt.X("year:O"),
    y=alt.Y("model", title='Model count', scale = alt.Scale(domain = [0, df_transmission.model.max()])),
    color=alt.Color("trany:N", title="Transmission"),
).properties(
   width=800
)

points = line.mark_point().encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

text = line.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(nearest, 'model', alt.value(' '))
)

rules = alt.Chart(df_transmission).mark_rule(color='gray').encode(
    x='year:O',
).transform_filter(
    nearest
)

selectors = alt.Chart(df_transmission).mark_point().encode(
    x='year:O',
    opacity=alt.value(0),
).add_selection(
    nearest
)

alt.layer(
    line, selectors, points, rules, text
)
```

Даний графік, на мою думку, дуже цікавий. Видно, що ще з початку (на моє превелике здивування, оскільки автомат в ті часи і ще довгі роки після того однозначно уступав механіці) кількість автомобілів на автоматі була більша, та йшла нога в ногу з механікою. Проте, саме десь в 2005, ці дві лінії почали віддалятися одна від одної: автомат - збільшуватись, механіка - вимирати. Зараз автомобілів на механіці випускають найменше за весь час представлений на графіку.

```python
df_fuel_cost = df.drop(df.columns.difference(['make', 'model', 'barrels08', 'barrelsA08', 'city08', 'cityA08', 'fuelCost08', 'fuelCostA08', 'co2TailpipeAGpm', 'co2TailpipeGpm', 'year']), 1)
df_fuel_cost
```

```python
df_fuel_cost = df_fuel_cost.groupby(["year"]).apply(lambda x: x.sort_values(["fuelCost08"], ascending = False)).reset_index(drop=True)
df_fuel_cost = df_fuel_cost.groupby('year').head(10)
df_fuel_cost
```

```python
slider = alt.binding_range(min=df_fuel_cost.year.min(), max=df_fuel_cost.year.max(), step=1, name='Year')
selector = alt.selection_single(name="YearSelector", fields=['year'],
                                bind=slider, init={'year': 2021})

alt.Chart(df_fuel_cost).mark_bar().encode(
    x=alt.X("model", title=None, sort=alt.SortField(field='fuelCost08'), axis=alt.Axis(labels=False, tickSize=0)),
    y=alt.Y("fuelCost08", title='Fuel cost', axis = alt.Axis(values=[df_fuel_cost.fuelCost08.min(), 4650, 5050, 5500,6050, 6750, df_fuel_cost.fuelCost08.max()]), scale = alt.Scale(domain = [df_fuel_cost.fuelCost08.min(), df_fuel_cost.fuelCost08.max()])),
    color=alt.Color("model"),
    tooltip=['make', 'model', 'fuelCost08']
).properties(
   width=800
).add_selection(
    selector
).transform_filter(
    selector
).resolve_scale( y='shared'
)
```

```python
df_drive = df.drop(df.columns.difference(['make', 'model',  'VClass', 'year']), 1)
df_drive['VClass'] = df_drive['VClass'].map(lambda x: x.rstrip(' -/24WDwd'))
# df_drive['VClass'] = df_drive['VClass'].str.replace('(\w+ Cars$)', 'Cars')
df_drive['VClass'] = df_drive['VClass'].str.replace('\w*(\-+|\ )Station Wagons', 'Station Wagons')
df_drive['VClass'] = df_drive['VClass'].str.replace('(\w+ Sport Utility Vehicle$)', 'Sport Utility Vehicle')
df_drive['VClass'] = df_drive['VClass'].str.replace('\w*Vans(,\ )*\w*\ \w*', 'Vans')
df_drive['VClass'] = df_drive['VClass'].str.replace('\w*(\-+|\ )Pickup Trucks', 'Pickup Trucks')
df_drive['VClass'] = df_drive['VClass'].str.replace('Vehicles$', 'Vehicle')
df_drive
df_drive
```

```python
slider = alt.binding_range(min=df_fuel_cost.year.min(), max=df_fuel_cost.year.max(), step=1, name='Year')
selector = alt.selection_single(name="YearSelector", fields=['year'],
                                bind=slider, init={'year': 2021})

dropdown = alt.binding_select(options=sorted(df_drive.make.unique()))
selector_make = alt.selection_single(name='Make', fields=['make'],
                                bind=dropdown, init={'make': 'Audi'})

pie = alt.Chart(df_drive).mark_arc(innerRadius=50,).encode(
    theta=alt.Theta(field="model", type="quantitative"),
    color=alt.Color(field="VClass", type="nominal", legend=alt.Legend(
        orient='top',
        direction='horizontal',
        titleAnchor='middle')),
).add_selection(
    selector
).add_selection(
    selector_make
).transform_filter(
    selector
).transform_filter(
    selector_make
).properties(
   width=500)

bar = alt.Chart(df_drive).mark_bar().encode(
    x=alt.X("VClass", title=None, axis=alt.Axis(labels=False, tickSize=0)),
    y=alt.Y("model", title=None, scale = alt.Scale(domain = [0, 22])),
    color=alt.Color("VClass"),
    tooltip=['make', 'model']
).properties(
   width=500)

alt.hconcat(
    pie, bar
).add_selection(
    selector
).add_selection(
    selector_make
).transform_filter(
    selector
).transform_filter(
    selector_make
)
```
